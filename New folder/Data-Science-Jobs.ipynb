```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# List of categorical columns to encode
categorical_columns = ['Company Name', 'Position','Job Title']

# Apply label encoding
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Display the updated dataset
print("Updated DataFrame with Label Encoding:")
print(df)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import re


# Function to clean and convert Salary column
def process_salary(salary):
    if pd.isna(salary):
        return np.nan
    # Extract numbers using regex
    numbers = re.findall(r'(\d+\.?\d*)', salary)
    # If there are numbers, convert to float and return the average of the range
    if len(numbers) == 2:
        low, high = float(numbers[0]), float(numbers[1])
        return (low + high) / 2
    elif len(numbers) == 1:
        return float(numbers[0])
    else:
        return np.nan

# Apply the function to the Salary column
df['Salary'] = df['Salary'].apply(process_salary)

# Function to convert Date column to numerical representation
def process_date(date):
    if pd.isna(date):
        return np.nan
    # Extract the number of days
    match = re.match(r'(\d+)d', date)
    return int(match.group(1)) if match else np.nan

# Apply the function to the Date column
df['Date'] = df['Date'].apply(process_date)
```
